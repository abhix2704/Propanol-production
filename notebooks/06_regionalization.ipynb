```python
import bw2analyzer as ba
import bw2data as bd
import bw2calc as bc
import bw2io as bi
import matrix_utils as mu
import bw_processing as bp
import time
import wurst
import os
from pathlib import Path
import copy
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
bd.projects.set_current('propanol_production')
```

```python
bd.databases
```

```python
db = bd.Database('inventories-ei310-Base-2025')
```

```python
method = ('IPCC 2021', 'climate change', 'GWP 100a, incl. H and bio CO2')
method
```

```python
df_2020 = pd.DataFrame(columns=['activity', 'location', 'score'])
```

```python
activity_names = [activity['name'] for activity in db if 'propanol production' in activity['reference product'] and 'GLO' in activity['location']]
activity_names
```

```python
activities = []
for ds in db:
    for activity_name in activity_names:
        if activity_name == ds['name']:
            activities.append(ds)
            break

for act in activities: 
    print(act['name'])
    lca = bc.LCA({act: 1}, method)
    lca.lci()
    lca.lcia()
    
    new_row = pd.DataFrame([{
    'activity': act['name'],
    'location': act['location'],
    'score': lca.score
    }])

    df_2020 = pd.concat([df_2020, new_row])


# Reset index of the DataFrame
df_2020.reset_index(drop=True, inplace=True)
df_2020.to_excel(os.path.join('..', 'results','regionalization_climate_change_2025.xlsx'))
```
